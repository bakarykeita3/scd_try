```python
import sympy as sp
from IPython.display import display

x, t = sp.symbols("x t")
T = sp.Function("T")(x, t)
eta = sp.symbols("eta")
#display(sp.diff(T, x))
#display(sp.diff(T, t))
#sp.Eq(sp.diff(T, t), eta * sp.diff(T, x, 2))

display(sp.Eq(sp.diff(T, t), eta * sp.diff(T, x, 2)))
#display(sp.series(T, x, 0, 3))
#display(sp.series(T, t, 0, 3))

"""je choisi la methode d'Euler centĆ©e pour la resolution de l'Ć©quation de la chaleur, oĆ¹ h est le pas de discrĆ©tisation spatiale 
et dt est le pas de discrĆ©tisation temporelle. On note T_i^n la valeur de T au point i et au temps n, 
et T_i^{n+1} la valeur de T au point i et au temps n+1. On a alors l'approximation suivante :"""

#discretisation spartiale
Ti_n = sp.symbols("T_i^n")
Ti_n1 = sp.symbols("T_{i+1}^n")
Ti_n2 = sp.symbols("T_{i-1}^n")
dx = sp.symbols("dx")

#discretisation temporelle
dt = sp.symbols("dt")
Ti_m1 = sp.symbols("T_i^{n+1}")
Ti_m2 = sp.symbols("T_i^n") 

# le nombre d'onde k et la pulsation omega.
k = sp.symbols("k")
#omega = sp.symbols("omega")

display(sp.Eq(sp.diff(T, x, 2), (Ti_n1 - 2*Ti_n + Ti_n2) / (2*dx**2)))
display(sp.Eq(sp.diff(T, t), (Ti_m1 - Ti_m2) / (2*dt)))

"""maintenant, on peut substituer ces approximations dans l'Ć©quation de la chaleur pour obtenir une Ć©quation discrĆØte qui peut ĆŖtre 
rĆ©solue numĆ©riquement."""
expr_discrete = sp.Eq((Ti_m1 - Ti_m2) / (2*dt), eta * ((Ti_n1 - 2*Ti_n + Ti_n2) / (2*dx**2)))
display(expr_discrete)



"""Afin de faciliter la rĆ©solution numĆ©rique, nous pouvons rĆ©Ć©crire cette Ć©quation sous forme matricielle. Mais avant cela,
on doit trouver la transformation de Fourier de l'Ć©quation discrĆØte. Pour ce faire, on peut utiliser
 la fonction sp.fourier_transform() de SymPy. Il facile de trouver cette transoformation. Par exemple,T_i^{n+/-1}=T_i^{n}*exp(+/-i*k*dx) 
 oĆ¹ k est le nombre d'onde. On peut donc Ć©crire l'Ć©quation discrĆØte sous forme matricielle en utilisant la transformation de Fourier."""

sp.Eq(Ti_n1, Ti_n * sp.exp(sp.I * k * dx))
sp.Eq(Ti_n2, Ti_n * sp.exp(-sp.I * k * dx))
#sp.Eq(Ti_m1, Ti_m2 * sp.exp(sp.I * omega * dt)) dans le cas d'un schema centrĆ© pour le temps.
sp.Eq(Ti_n, Ti_n)
expr_discrete_fourier = expr_discrete.subs({Ti_n1: Ti_n*sp.exp(sp.I*k*dx), Ti_n2: Ti_n*sp.exp(-sp.I*k*dx), Ti_n: Ti_n})
display(expr_discrete_fourier)
```
