```python
import sympy as sp
from IPython.display import display

x, t = sp.symbols("x t")
T = sp.Function("T")(x, t)
eta = sp.symbols("eta")
#display(sp.diff(T, x))
#display(sp.diff(T, t))
#sp.Eq(sp.diff(T, t), eta * sp.diff(T, x, 2))

display(sp.Eq(sp.diff(T, t), eta * sp.diff(T, x, 2)))
#display(sp.series(T, x, 0, 3))
#display(sp.series(T, t, 0, 3))

"""je choisi la methode d'Euler centée pour la resolution de l'équation de la chaleur, où h est le pas de discrétisation spatiale 
et dt est le pas de discrétisation temporelle. On note T_i^n la valeur de T au point i et au temps n, 
et T_i^{n+1} la valeur de T au point i et au temps n+1. On a alors l'approximation suivante :"""

#discretisation spartiale
Ti_n = sp.symbols("T_i^n")
Ti_n1 = sp.symbols("T_{i+1}^n")
Ti_n2 = sp.symbols("T_{i-1}^n")
dx = sp.symbols("dx")

#discretisation temporelle
dt = sp.symbols("dt")
Ti_m1 = sp.symbols("T_i^{n+1}")
Ti_m2 = sp.symbols("T_i^n") 

# le nombre d'onde k et la pulsation omega.
k = sp.symbols("k")
#omega = sp.symbols("omega")

display(sp.Eq(sp.diff(T, x, 2), (Ti_n1 - 2*Ti_n + Ti_n2) / (2*dx**2)))
display(sp.Eq(sp.diff(T, t), (Ti_m1 - Ti_m2) / (2*dt)))

"""maintenant, on peut substituer ces approximations dans l'équation de la chaleur pour obtenir une équation discrète qui peut être 
résolue numériquement."""
expr_discrete = sp.Eq((Ti_m1 - Ti_m2) / (2*dt), eta * ((Ti_n1 - 2*Ti_n + Ti_n2) / (2*dx**2)))
display(expr_discrete)



"""Afin de faciliter la résolution numérique, nous pouvons réécrire cette équation sous forme matricielle. Mais avant cela,
on doit trouver la transformation de Fourier de l'équation discrète. Pour ce faire, on peut utiliser
 la fonction sp.fourier_transform() de SymPy. Il facile de trouver cette transoformation. Par exemple,T_i^{n+/-1}=T_i^{n}*exp(+/-i*k*dx) 
 où k est le nombre d'onde. On peut donc écrire l'équation discrète sous forme matricielle en utilisant la transformation de Fourier."""

sp.Eq(Ti_n1, Ti_n * sp.exp(sp.I * k * dx))
sp.Eq(Ti_n2, Ti_n * sp.exp(-sp.I * k * dx))
#sp.Eq(Ti_m1, Ti_m2 * sp.exp(sp.I * omega * dt)) dans le cas d'un schema centré pour le temps.
sp.Eq(Ti_n, Ti_n)
expr_discrete_fourier = expr_discrete.subs({Ti_n1: Ti_n*sp.exp(sp.I*k*dx), Ti_n2: Ti_n*sp.exp(-sp.I*k*dx), Ti_n: Ti_n})
display(expr_discrete_fourier)
```

```python
lhs_factored = sp.factor(Ti_m1)
rhs_factored = sp.factor(expr_discrete_fourier)
equation = sp.Eq(lhs_factored, rhs_factored)
display(equation)

```
